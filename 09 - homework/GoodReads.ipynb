```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors')
doc = BeautifulSoup(response.text)
```

```python
ranks = doc.find_all(class_='minirating')
for rank in ranks:
    print(rank)
```

```python
titles = doc.find_all(class_='bookTitle')
for title in titles:
    print(title)
```

```python
authors = doc.find_all(itemprop="author")
for author in authors:
    print(author)
```

```python
scores = doc.find_all(class_='smallText uitext')
for score in scores:
    print(score)
```

```python
votes = doc.find_all(a_id ="loading_link_566843")
for vote in votes:
    print(vote)    
```

```python
ratings = doc.find_all(class_="minirating")
for rating in ratings:
    print(rating) 
```

```python
url = "https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors"
print("I am scraping", url)

response = requests.get(url)
doc = BeautifulSoup(response.text)

good_reads = {}

good_reads['ratings'] = doc.find_all(class_="minirating")
good_reads['votes'] = doc.find_all(a_id ="loading_link_566843")
good_reads['scores'] = doc.find_all(class_='smallText uitext')
good_reads['ranks'] = doc.find_all(class_='minirating')
good_reads['authors'] = doc.find_all(itemprop="author")
good_reads['titles'] = doc.find_all(class_='bookTitle')
print(good_reads).string()
```
