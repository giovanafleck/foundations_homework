```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

response = requests.get('https://www.billboard.com/charts/hot-100')
doc = BeautifulSoup(response.text)
```

Scrape the fields below, and save as a CSV file.

Rank;
Artist name;
Song name;

```python
ranks = doc.find_all(class_='chart-list-item__rank')
for rank in ranks:
    print(rank.text)
```

```python
artists = doc.find_all(class_='chart-list-item__artist')
for artist in artists:
    print(artist.text)
```

```python
songs = doc.find_all(class_='chart-list-item__title-text')
for song in songs:
    print(song.text)
```

```python
url = "https://www.billboard.com/charts/hot-100"
print("I am scraping", url)

response = requests.get(url)
doc = BeautifulSoup(response.text)

billboard = {}

billboard['songs'] = doc.find_all(class_='chart-list-item__title-text')
billboard['artists'] = doc.find_all(class_='chart-list-item__artist')
billboard['ranks'] = doc.find_all(class_='chart-list-item__rank')
print(billboard)
```

```python
df = pd.DataFrame(billboard)
df.head()
```

```python
df.to_csv("billboard.csv", index=False)
```
